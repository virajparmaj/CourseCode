**Part 1-1**  
Simulate a Random Walk $X_t=X_{t-1}-0.1+W_t$, where $\{W_t\}_{t\geq 0}$
is a white noise with with standard deviation 1.2. Include your python
code and plots generated from python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 100
std_dev = 1.2
np.random.seed(30)

# random walk
W_t = np.random.normal(loc=0, scale=std_dev, size=n) 
X_t = np.zeros(n) 
X_t[0] = 0 

for t in range(1, n):
    X_t[t] = X_t[t-1] - 0.1 + W_t[t]

# Plotting the random walk
plt.figure(figsize=(10, 6))
plt.plot(X_t, label='Random Walk', color='blue')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.8)
plt.title("Simulated Random Walk: $X_t = X_{t-1} - 0.1 + W_t$")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = "GME_stock.csv"

# Load the data and parse the date column
gme_data = pd.read_csv(file_path, parse_dates=["date"], index_col="date")

# 5 observations
print("First 5 Observations:")
print(gme_data.head())

# time series plot 'close_price'
plt.figure(figsize=(12, 6))
plt.plot(gme_data['close_price'], label="Close Price", color='blue')
plt.title("GameStop Stock Close Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price (USD)")
plt.legend()
plt.grid()
plt.show()
```
